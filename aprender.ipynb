```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_alumnos = pd.read_csv("csv_files_app_alumno.csv")
```

```python
df_escuelas = pd.read_csv("csv_files_app_alumno.csv")
```

```python
object_columns = ['sexo','indice_socioeconomico','nivel_desemp_matematica','nivel_desemp_lengua','nivel_desemp_ciencias_sociales','nivel_desemp_ciencias_naturales','tiene_notebook','tiene_pc','tiene_tablet','tiene_celular','tiene_smartphone','tiene_consola','tiene_smarttv','tiene_cable','tiene_internet','repeticion_primaria','repeticion_secundaria']
```

```python
for x in object_columns:
    print(df_alumnos[x].value_counts())
    print()
```

```python
df_alumnos.dtypes
```

```python
for col_name in df_alumnos.columns:
    if(df_alumnos[col_name].dtype == 'object'):
        print(df_alumnos[col_name].value_counts(ascending=True, normalize=True, sort=True))
        plt.figure()
        df_alumnos[col_name].value_counts(ascending=True, normalize=True, sort=True).plot(kind='bar', title=col_name)
        print()
```

```python
for x in object_columns:
    df_alumnos[x].astype('category')
```

```python
df_alumnos.corr()
```

```python
for col_name in df_alumnos.columns:
    if(df_alumnos[col_name].dtype == 'object'):
        df_alumnos[col_name]= df_alumnos[col_name].astype('category')
        df_alumnos[col_name] = df_alumnos[col_name].cat.codes
```
